# Student Loan Risk with Deep Learning

```python
# Imports
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from pathlib import Path
```

---

## Prepare the data to be used on a neural network model

### Step 1: Read the `student-loans.csv` file into a Pandas DataFrame. Review the DataFrame, looking for columns that could eventually define your features and target variables.   

```python
# Read the csv into a Pandas DataFrame
file_path = "https://static.bc-edx.com/ai/ail-v-1-0/m18/lms/datasets/student-loans.csv"
loans_df = pd.read_csv(file_path)

# Review the DataFrame
loans_df.head()
```

```python
# Review the data types associated with the columns
loans_df.dtypes
```

```python
# Check the credit_ranking value counts
loans_df["credit_ranking"].value_counts()
```

### Step 2: Using the preprocessed data, create the features (`X`) and target (`y`) datasets. The target dataset should be defined by the preprocessed DataFrame column “credit_ranking”. The remaining columns should define the features dataset.

> Add blockquote



```python
# Define the target set y using the credit_ranking column
y = loans_df["credit_ranking"]

# Display a sample of y
y.values
```

```python
# Define features set X by selecting all columns but credit_ranking
X = loans_df.drop(columns=['credit_ranking'])

# Review the features DataFrame
X.head()
```

### Step 3: Split the features and target sets into training and testing datasets.


```python
# Split the preprocessed data into a training and testing dataset
# Assign the function a random_state equal to 1
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

### Step 4: Use scikit-learn's `StandardScaler` to scale the features data.

```python
# Create a StandardScaler instance
X_scaler = StandardScaler()


# Fit the scaler to the features training dataset
X_scaler.fit(X_train)


# Fit the scaler to the features training dataset
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

---

## Compile and Evaluate a Model Using a Neural Network

### Step 1: Create a deep neural network by assigning the number of input features, the number of layers, and the number of neurons on each layer using Tensorflow’s Keras.

> **Hint** You can start with a two-layer deep neural network model that uses the `relu` activation function for both layers.


```python
# Define the the number of inputs (features) to the model
input_nodes = len(X.columns)


# Review the number of features
input_nodes
```

```python
# Define the number of hidden nodes for the first hidden layer
first_hidden_nodes = 6


# Review the number hidden nodes in the first layer
print (first_hidden_nodes)

# Define the number of hidden nodes for the second hidden layer
second_hidden_nodes = 3


# Review the number hidden nodes in the second layer
print (second_hidden_nodes)

# Define the number of neurons in the output layer
number_of_neurons = 1
```

```python
# Create the Sequential model instance
model = tf.keras.models.Sequential()

# Add the first hidden layer
model.add(tf.keras.layers.Dense(first_hidden_nodes, activation='relu', input_dim=input_nodes))

# Add the second hidden layer
model.add(tf.keras.layers.Dense(second_hidden_nodes, activation='relu'))


# Add the output layer to the model specifying the number of output neurons and activation function
model.add(tf.keras.layers.Dense(number_of_neurons, activation='sigmoid'))
```

```python
# Display the Sequential model summary
model.summary()
```

### Step 2: Compile and fit the model using the `binary_crossentropy` loss function, the `adam` optimizer, and the `accuracy` evaluation metric.


```python
# Compile the Sequential model
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

```

```python
# Fit the model using 50 epochs and the training data
fit_model = model.fit(X_train_scaled, y_train, epochs=50)
```

### Step 3: Evaluate the model using the test data to determine the model’s loss and accuracy.


```python
# Evaluate the model loss and accuracy metrics using the evaluate method and the test data
model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test, verbose=2)

# Display the model loss and accuracy results
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

**1. Describe the data that you would need to collect to build a recommendation system to recommend student loan options for students. Explain why this data would be relevant and appropriate.**

To build an effective recommendation system for student loans, the following data would be crucial:

- **Student Demographics and Background**: Age, location, education level, field of study, institution, academic performance (GPA), graduation date, career aspirations, and financial background (income, existing debt).
- **Loan Characteristics**: Interest rates, loan terms (length of repayment), loan amounts, fees, lender type (federal, private), repayment options, and eligibility criteria.
- Student Preferences: Preferred loan features (e.g., low interest rates, flexible repayment options), risk tolerance, and financial goals.
- Loan Performance Data: Historical data on loan repayment behavior, default rates, and factors associated with successful loan repayment.

**Relevance and Appropriateness**:

This comprehensive data provides a holistic view of the student's financial situation, academic profile, and loan preferences. By analyzing this data, the recommendation system can identify suitable loan options aligning with the student's individual needs and goals, minimizing the risk of default and maximizing the likelihood of successful repayment.


**2. Based on the data you chose to use in this recommendation system, would your model be using collaborative filtering, content-based filtering, or context-based filtering? Justify why the data you selected would be suitable for your choice of filtering method.**

Based on the data described above, a context-based filtering approach would be most suitable for this recommendation system. Context-based filtering considers the user's current context and situation to provide personalized recommendations. In this case, the student's financial background, academic progress, career aspirations, and loan preferences form the context.

**Justification**:

By incorporating contextual information, the recommendation system can provide more relevant and tailored loan suggestions. For example:

- A student with a high GPA and a promising career path might be recommended loans with lower interest rates and longer repayment terms.
- A student with existing debt might be recommended loans with income-based repayment options 

**3. Describe two real-world challenges that you would take into consideration while building a recommendation system for student loans. Explain why these challenges would be of concern for a student loan recommendation system.**

Two critical real-world challenges to consider when building a student loan recommendation system are:

- Data Privacy and Security: Protecting sensitive student data is paramount. The system must comply with relevant privacy regulations (e.g., FERPA) and implement robust security measures to safeguard student information.
- Bias and Fairness: The recommendation system should avoid perpetuating existing biases or discriminating against certain groups of students. Careful consideration must be given to ensure that the system's algorithms and data sources are fair and unbiased.

**Why These Challenges are Important**:

- Data privacy breaches could expose sensitive student information, leading to financial harm and reputational damage.
- Biased recommendations could unfairly limit opportunities for certain students, exacerbating existing inequalities in access to education and financial resources.
